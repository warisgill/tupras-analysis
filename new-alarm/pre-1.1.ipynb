### Summary
Concatination of 3 csv files into 1. 

```python
import numpy as np
import json
import pandas as pd
from datetime import datetime
from dateutil.parser import parse
```

```python
f1 = "formatted-pre-1-nisan2019.csv"
f2 = "formatted-pre-1-mayis2019.csv" 
f3 = "formatted-pre-1-haziran2019.csv"
output_f = 'formatted-pre-1-months3.csv' 

path = "./data/new/"
df1 = pd.read_csv(path + f1, parse_dates = ["EventTime"])
df2 = pd.read_csv(path + f2, parse_dates = ["EventTime"])
df3 = pd.read_csv(path + f3, parse_dates = ["EventTime"])
```

```python
df = pd.concat([df1, df2, df3], ignore_index=True)
df.to_csv( path + output_f, index=False)
df
```

```python
cols = ["MachineName","SourceName","EventTime", "Message","Quality","Condition","Mask","NewState","Status"]
df = pd.read_csv(path + output_f, usecols=cols ,parse_dates = ["EventTime"], low_memory=False)
df.to_csv( path + output_f, index=False)
df
```
